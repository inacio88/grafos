Algoritmo 2: O funcionamento desse algoritmo é um pouco mais complexo do que o anterior; aqui temos como entrada a matriz de adjacência, e precisamos retornar a quantidade de elementos conexos. O primeiro passo consiste em transformarmos o grafo num grafo simples, e isso é feito eliminando-se os loops (colocando 0 na diagonal principal) e arestas paraleas (não permitir valores maiores que 1); o próximo passo é a fusão de vértices adjacentes e, logo após, verificar se todos os elementos da matriz são iguais a zero; caso isso não seja verificado, os passos anteriores são repetidos até que os elementos da matriz sejam iguais a zero, a quantidade de vértices que essa matriz final indicar será a quantidade de componentes conexos do grafo.

# Graus de cada vértice

```python
def soma_linha(matrizIncidencia):
    graus = []
    for linha in matrizIncidencia:
        grau = 0
        for elemento in linha:
            grau = grau + elemento
        graus.append(grau)
    tamanho = len(graus)
    for i in range(tamanho):
        print("O grau do vértice {} é {}".format(i, graus[i]))
        
        
def exibir_matriz(matriz):
    for linha in matriz:
        print(linha)

        
        
        
m = int(input("Número de vertices: "))
n = int(input("Número de arestas: "))

matrizIncidencia = []
for i in range(m):
    linha = []
    for j in range(n):
        elemento = int(input("Elemento da linha {} e coluna {} ".format(i,j)))
        linha.append(int(elemento))
    matrizIncidencia.append(linha)

    
    
exibir_matriz(matrizIncidencia)
soma_linha(matrizIncidencia)
```
